```python
import nylib
import hzlib
%pylab inline
reload(hzlib)
```

```python
# Define process parameter
dataPath = '/Users/hzeng/Data/140414_folic_acid/34'
vdListFile = loadtxt('%s/vdlist'%dataPath)

TD = int(nylib.BrukerPar(dataPath, 'acqu2', 'TD'))
data = nylib.BrukerFid(dataPath)
tevol = loadtxt(vdListFile)[:TD]
data = data.reshape((TD,len(data)/TD))
```

```python
ExampleProjection = abs(fft.fftshift(fft.fft(data[0])))
f, (axspec, axfid) = subplots(2, 1)
axspec.plot(real(data[0]))
axfid.plot(ExampleProjection)
flexFid = hzlib.fftFlex(data)
```

```python
# fit out baseline
p0 = array( [0.1, 0.7, 10., 1., 500., 0.1, 1000., 1., 500.])
p1 = nylib.FlexLSFit(tevol, flexFid, p0)
nylib.PrintResults(p1)
baseline = nylib.fitfunc(r_[p1[0], p1[1:5]], tevol)
fidFiltered = flexFid - baseline

f, (axa, axb) = subplots(2, 1)
axa.plot(tevol*1e3, flexFid, 'b')
axa.plot(tevol*1e3, nylib.fitfunc(p1, tevol)-0.1,'g')
axa.plot(tevol*1e3, baseline-0.2, 'r')
axb.plot(tevol*1e3, flexFid-nylib.fitfunc(p1, tevol), 'y')
axb.plot(tevol*1e3, fidFiltered)
```

```python
# transform to Fourier domain
phase0 = 3.8
phase1 = 0.0
fill = 8192*2
apodize = 1.
o1 = float(nylib.BrukerPar(dataPath, 'acqu', 'O2')) - float(nylib.BrukerPar(dataPath, 'acqu', 'O1'))
#fid2, FlexSpectrum = nylib.getSpectrum((fidFiltered - nylib.fitfunc(r_[0, p1[1:5]], tevol)), tevol, o1, 0.0, 0., 8192*2, 1)
fid2, flexSpec = hzlib.getFlexSpec(fidFiltered, tevol, o1, phase0, phase1, fill, apodize)

ppm = nylib.PpmScale(flexSpec, o1, 750, tevol[2]-tevol[1], '1D')
```
